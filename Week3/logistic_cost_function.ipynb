```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_train = np.array([[0.5, 1.5], [1, 1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y_train = np.array([0, 0, 0, 1, 1, 1])
# m = 6
# n = 2
```

```python
fig, ax = plt.subplots(1, 1, figsize=(4, 4))
pos = y_train==1
neg = y_train==0
ax.scatter(X_train[:,0][pos], X_train[:,1][pos], marker="x", color="r")
ax.scatter(X_train[:,0][neg], X_train[:,1][neg], marker="o", color="b")

ax.set_ylabel("$x_1$", fontsize=12)
ax.set_xlabel("$x_0$", fontsize=12)
plt.show()
```

```python
def sigmoid(z):
  g = 1/(1+np.exp(-z))
  return g
```

```python
def compute_cost_logistic(X, y, w, b):
    """
    Computes cost

    Args:
      X (ndarray (m,n)): Data, m examples with n features
      y (ndarray (m,)) : target values
      w (ndarray (n,)) : model parameters
      b (scalar)       : model parameter

    Returns:
      cost (scalar): cost
    """
    m = X.shape[0]
    cost=0
    for i in range(m):
      z_i = np.dot(X[i], w)+b
      f_wb_i = sigmoid(z_i)
      loss_i = -y[i]*np.log(f_wb_i)-(1-y[i])*np.log(1-f_wb_i)
      cost+=loss_i
    cost = cost/m
    return cost
```

```python
# checking above computing cost func
w_tmp = np.array([1, 1])
b_tmp=-3
print(compute_cost_logistic(X_train, y_train, w_tmp, b_tmp))
```

```python
# choose values between 0 and 6
x0=np.arange(0, 6)

# plot the two decision boundaries
x1 = 3-x0
x1_other = 4-x0

fig,ax = plt.subplots(1, 1, figsize=(4, 4))
# Plot the decision boundary
ax.plot(x0, x1, c="b", label="$b$=-3")
ax.plot(x0, x1_other, c="m", label="$b$=-4")
ax.axis([0, 4, 0, 4])

# Plot the original data
ax.scatter(X_train[:,0][pos], X_train[:,1][pos], marker="x", color="r")
ax.scatter(X_train[:,0][neg], X_train[:,1][neg], marker="o", color="b")
ax.axis([0, 4, 0, 4])
ax.set_ylabel('$x_1$', fontsize=12)
ax.set_xlabel('$x_0$', fontsize=12)
plt.legend(loc="upper right")
plt.title("Decision Boundary")
plt.show()
```

```python
# above magenta decision boundary is worser than blue, we can check this by
# calculating the cost
b_1=-3
b_2=-4

print("Cost for b = -3 : ", compute_cost_logistic(X_train, y_train, w_tmp, b_1))
print("Cost for b = -4 : ", compute_cost_logistic(X_train, y_train, w_tmp, b_2))
```
